Partitioning the data according to the species field

```python
setosa=data.loc[data['Species']=='Iris-setosa']
setosa.describe()
```

**2-D Scatter Plot**

The main idea behind this is to plot a scatter plot by taking 2 featues at a time then after that identifying if we are able to differentiate the three classes of flower present in the dataset

```python
plt.figure(figsize=(11,6))
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=data)

plt.show()
```

```python
plt.figure(figsize=(11,6))
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=data.loc[data['Species']=='Iris-setosa'],color='blue',label='setosa')
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=data.loc[data['Species']=='Iris-versicolor'],color='green',label='versicolor')
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=data.loc[data['Species']=='Iris-virginica'],color='red',label='virginica')
plt.legend(loc=4)
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('Species Distribution')
plt.show()
```

**Pair-Plot**

This feature is generally used only when the number of features in the dataset is high

```python

sns.set_style('whitegrid')
sns.pairplot(data,hue='Species',height=4)
plt.show()
```

**Uni-Variate Analysis**

1-D Scatter Plot of Petal Length

```python
plt.figure(figsize=(11,6))
sns.FacetGrid(data, hue="Species") \
   .map(sns.distplot, "PetalLengthCm") \
   .add_legend()
plt.title('Petal Length Analysis')
plt.show()
```

1-D Scatter Plot of Petal Width

```python
sns.FacetGrid(data, hue="Species") \
   .map(sns.distplot, "PetalWidthCm") \
   .add_legend()
plt.title("Petal Width Analysis")
plt.show()
```

1-D Scatter Plot of Sepal Width

```python
sns.FacetGrid(data, hue="Species") \
   .map(sns.distplot, "SepalWidthCm") \
   .add_legend()
plt.title("Sepal Width Analysis")
plt.show()
```

1-D Scatter Plot of Sepal Length

```python
sns.FacetGrid(data, hue="Species") \
   .map(sns.distplot, "SepalLengthCm") \
   .add_legend()
plt.title("Sepal Length Analysis")
plt.show()
```

**Observation**

After performing a 1-D Scatter plot it is clearly visible that the data points are overlapping and that it is difficult to separate the points

Thus we can conclude that:
1. By performing histogram plot, it is clear that sepal_length and sepal width feature have many data points overlapping so it is difficult to build a model by taking these two feature.
2. Petal_length and petal_width is more beneficial while taking a plot
3. PL>PW>SL>SW this is the order which can be taken into consideration while classifying a dataset

```python
sns.boxplot(x='Species',y='PetalLengthCm', data=data)
plt.show()
```

Finding the mean values of the Petal Length Feature in the order of Iris-Setosa,Iris-Versicolor and Iris-Virginica respectively.

```python
print("Means:")
print(np.mean(setosa["PetalLengthCm"]))
#Mean with an outlier.
print(np.mean(np.append(setosa["PetalLengthCm"],50)));
print(np.mean(virginica["PetalLengthCm"]))
print(np.mean(versicolor["PetalLengthCm"]))

print("Medians:")
print(np.median(setosa['PetalLengthCm']))
print(np.median(virginica["PetalLengthCm"]))
print(np.median(versicolor["PetalLengthCm"]))


print("\nStd-dev:");
print(np.std(setosa["PetalLengthCm"]))
print(np.std(virginica["PetalLengthCm"]))
print(np.std(versicolor["PetalLengthCm"]))
```
